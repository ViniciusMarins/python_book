```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cols = [
    'Entity', 'Code', 'Year', 'Alochol use',
    'Diet low in fruits', 'Unsafe water source','Secondhand smoke',
    'Smoking','Air pollution','High body mass index','Drug use'
]

df = pd.read_csv('dataset.csv', usecols=cols)

paises = ['Brazil', 'Pakistan', 'United States', 'Japan', 'Russia', 'Mexico', 'Ethiopia', 'Philippines','India','China']

dataframe = df[df['Entity'].isin(paises)]
```

```python
causas = list(filter(lambda col: col not in ['Entity', 'Code', 'Year'], cols))
```

```python
paises = dataframe['Entity'].unique()
cores_paises = {pais: np.random.rand(3,) for pais in paises}

for causa in causas:
    plt.figure(figsize=(12, 6)) 
    
    for pais in paises:
        dataframe_pais = dataframe[dataframe['Entity'] == pais]
        plt.bar(pais, dataframe_pais[causa], label=pais, color=cores_paises[pais])
    
    plt.xlabel('Pas')
    plt.ylabel('Nmero de Mortes')
    plt.xticks(rotation=90)  
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    
    plt.grid(True)  
    plt.title(f'Nmero de Mortes por Causa: {causa}')
    plt.tight_layout() 
    plt.show()
```

```python
df_brasil = dataframe[dataframe['Entity'] == 'Brazil']

plt.figure(figsize=(12, 6))

for causa in causas:
    plt.plot(df_brasil['Year'], df_brasil[causa], label=causa)

plt.xlabel('Ano')
plt.ylabel('Nmero de Mortes')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.grid(False)  
plt.title('Nmero de Mortes por Causa no Brasil')
plt.show()
```
